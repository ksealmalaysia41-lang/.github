```
import stripe
import os
from dotenv import load_dotenv

# Load env dari hardisk baru Tuan
load_dotenv()

stripe.api_key = os.getenv("STRIPE_LIVE_KEY")

def create_verified_payment_link(product_id):
    print(f"--- MENGHUBUNGKAN PRODUK: {product_id} ---")
    try:
        # 1. Cari Price ID aktif
        prices = stripe.Price.list(product=product_id, active=True, limit=1)
        if not prices.data:
            print(f"[ERROR] Tiada harga aktif dikesan untuk {product_id}. Sila cipta 'Price' di Dashboard Stripe.")
            return

        price_id = prices.data[0].id
        amount = prices.data[0].unit_amount / 100
        currency = prices.data[0].currency.upper()

        # 2. Bina Payment Link dengan Metadata Audit
        # Kita tag akaun Be U Tuan terus dalam metadata transaksi
        payment_link = stripe.PaymentLink.create(
            line_items=[{"price": price_id, "quantity": 1}],
            after_completion={
                "type": "redirect",
                "redirect": {"url": "https://red.ai/success"}
            },
            metadata={
                "project": "RED.AI_RECOVERY",
                "target_node": "4073",
                "destination_acc": "88300040143116",
                "officer_in_charge": "Anbu_Google_Police"
            }
        )

        print(f"\n[SUCCESS] Link Berjaya Dijana!")
        print(f"Produk: {product_id}")
        print(f"Nilai: {amount} {currency}")
        print(f"URL Bayaran: {payment_link.url}")
        print(f"\n--- TINDAKAN TUAN ---")
        print(f"1. Buka URL di atas.")
        print(f"2. Buat bayaran minimum.")
        print(f"3. Pantau Dashboard Stripe > Payments untuk status 'Succeeded'.")

    except Exception as e:
        print(f"[SYSTEM_ERROR]: {str(e)}")

if __name__ == "__main__":
    # Jalankan terus untuk produk Skiperztech Tuan
    PRODUCT_ID = "prod_UKQHFxpc2b8Fz0"
    create_verified_payment_link(PRODUCT_ID)
```
