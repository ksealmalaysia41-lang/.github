```
import signal
import time
import os
import subprocess
import sys

def sync_to_github(message="Auto-sync Redeye V7"):
    """Fungsi mekanikal untuk push kod ke GitHub"""
    try:
        # Check jika ada perubahan
        status = subprocess.check_output(["git", "status", "--porcelain"]).decode("utf-8")
        if status:
            print(f"[GITHUB] Perubahan dikesan pada {time.strftime('%H:%M:%S')}. Memulakan sinkronisasi...")
            subprocess.run(["git", "add", "."], check=True)
            subprocess.run(["git", "commit", "-m", message], check=True)
            # Menggunakan push tanpa prompting jika SSH sudah di-set
            subprocess.run(["git", "push"], check=True)
            print("[GITHUB] Sync berjaya.")
        else:
            print("[GITHUB] Tiada perubahan kod dikesan. Integriti selamat.")
    except Exception as e:
        print(f"[GITHUB ERROR] Gagal sync: {e}")

def handle_chakra_interruption(signum, frame):
    """Handler untuk shutdown tertib oleh Tini"""
    print(f"\n[!] Signal {signum} (Chakra Interruption) dikesan!")
    print("[*] Tini sedang mengemas peranti dan membersihkan sisa serangan...")
    print("[*] Melakukan final sync ke GitHub sebelum tamat...")

    # Final Emergency Sync
    sync_to_github("Emergency Shutdown Sync: Redeye V7 Termination")

    print("[✓] Benang chakra diputus dengan selamat. Re-routing ke Google Cloud Shell...")
    sys.exit(0)

# Daftar signal Tini (SIGTERM untuk Docker stop, SIGINT untuk Ctrl+C)
signal.signal(signal.SIGTERM, handle_chakra_interruption)
signal.signal(signal.SIGINT, handle_chakra_interruption)

def start_maraton():
    # Mengambil info dari ENV
    target_zone = os.getenv("TARGET_ZONE", "Cyberjaya Loop 101")
    project_id = "gen-lang-client-0644163502"

    print("--- REDEYE V7: KUGUTSU MODE ACTIVE ---")
    print(f"PID: {os.getpid()} | Managed by Tini (PID 1)")
    print(f"Agent: CISO Iskandar | Project: {project_id}")
    print(f"Target: {target_zone}")

    last_sync = time.time()

    try:
        while True:
            current_time = time.strftime('%H:%M:%S')
            print(f"[>] [{current_time}] Drone sedang aktif pusing bangunan {target_zone}...")

            # Simulasi logic monitor: Attacker Fatigue Check
            print("[!] Status: Attacker Mental Fatigue Level: INCREASING...")

            # Auto-sync setiap 10 minit (600 saat)
            if time.time() - last_sync > 600:
                sync_to_github(f"Periodic Sync - {current_time}")
                last_sync = time.time()

            # Interval pergerakan drone
            time.sleep(10)

    except Exception as e:
        print(f"[ERROR] Sistem mengalami kegagalan kritikal: {e}")
        sync_to_github(f"Crash Log Sync: {e}")

if __name__ == "__main__":
    start_maraton()
```
